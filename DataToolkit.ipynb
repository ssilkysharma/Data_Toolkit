2.	How does broadcasting work in NumPy?
Sol: Broadcasting in NumPy refers to the way NumPy handles element-wise binary operations (like +,-,*, etc.) when the two arrays involved do not have the same shape.
It allows NumPy to perform operations on arrays of different shapes in such a way that it "stretches" or "expands" the smaller array to match the shape of the larger array. This eliminates the need for explicit loops and improves performance.
Broadcasting Rules:
1.	If the arrays have a different number of dimensions, the shape of the smaller array is padded with ones on the left until both arrays have the same number of dimensions.
2.	If the dimensions are not the same, broadcasting happens when:
o	One of the arrays has a size of 1 along that dimension.
o	The other array's corresponding dimension has a size greater than 1.
3.	The arrays are compatible for broadcasting if, after applying the rules, their dimensions are either the same or the size of one of them is 1.


```python
import numpy as np

# A 2D array
A = np.array([[1, 2, 3],
              [4, 5, 6]])

# A 1D array
B = np.array([10, 20, 30])

# Broadcasting: Adding a 1D array to a 2D array
C = A + B

print(C)
```

4.	 Explain the use of the groupby() method in Pandas
Ans: In Pandas, the groupby() method is a powerful tool that allows you to split a DataFrame or Series into groups based on certain criteria, apply a function to each group independently, and then combine the results into a new DataFrame or Series. It's commonly used in data analysis tasks like aggregation, transformation, and filtering.
Syntax: df.groupby(by)


```python
import pandas as pd

# Sample DataFrame
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'East', 'South'],
    'Sales': [100, 200, 300, 400, 150, 350, 250]
}

df = pd.DataFrame(data)

# Group by the 'Region' column and calculate the total sales for each region
grouped = df.groupby('Region')['Sales'].sum()

print(grouped)
```

17.	Explain the difference between apply() and map() in Pandas
Ans: The apply() and map() functions in Pandas are both used to transform data, but they differ in how and where they operate. The map() function is specifically designed for use with Pandas Series objects and performs element-wise operations, meaning it applies a function, dictionary, or Series mapping to each value individually. It is best suited for simple transformations, such as replacing values or applying a function to each item in a single column.
In contrast, the apply() function is more versatile. It works with both Series and DataFrames, and its behavior depends on the context. When used with a Series, it behaves similarly to map() by applying a function to each element. However, when used on a DataFrame, apply() can operate either row-wise or column-wise, depending on the axis argument (axis=0 for columns, axis=1 for rows). This makes apply() suitable for more complex operations that involve multiple columns or rows.

21.Why is NumPy’s array slicing faster than Python’s list slicing
Ans: NumPy’s array slicing is faster than Python’s list slicing primarily because of how the two data structures are implemented and managed in memory.
NumPy arrays store elements in a contiguous block of memory with a fixed data type, which allows slicing to create views rather than copies. This means when you slice a NumPy array, it doesn't copy the data but creates a new array object that references the same underlying data in memory. This view is very fast to create and uses less memory.
On the other hand, Python lists are collections of pointers to separate Python objects scattered throughout memory. When you slice a list, Python has to create a new list object and copy the references (pointers) to the sliced elements into this new list. This copying process is slower and uses more memory.
In summary, NumPy's contiguous memory layout and slicing by view make array slicing much faster and more memory-efficient than Python's list slicing, which involves creating new lists and copying references.

4.	 How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'Math': [90, 85, 88, 92, 95],
    'Science': [85, 80, 86, 89, 94],
    'English': [78, 75, 80, 82, 85]
}

df = pd.DataFrame(data)

# Step 1: Calculate the correlation matrix
corr_matrix = df.corr()

# Step 2: Visualize with a heatmap
plt.figure(figsize=(4, 2))
sns.heatmap(corr_matrix, annot=True, cmap='viridis', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

13.	Create a 3D scatter plot using Plotly

```python
import plotly.graph_objects as go
import numpy as np

# Generate random 3D data
np.random.seed(24)
x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)

# Create 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=4,
        color=z,              # set color to an array/list of z values
        colorscale='Viridis', # choose a colorscale
        opacity=0.4
    )
)])

fig.update_layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    title='3D Scatter Plot Example'
)

fig.show()
```
